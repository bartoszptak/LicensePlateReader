```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
import glob

arrays = glob.glob('./data/cars/*.npy')
data_path = 'data'
div = 10

data_set = []
for path in arrays:
    local_set = np.load(os.path.join(path))
    for it in local_set:
        data_set.append(it)
data_set = np.array(data_set)
```

```python
for im in data_set:
    img = cv2.resize(im[0], (int(1280/div), int(720/div)))
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    im[0] = cv2.equalizeHist(gray)
```

```python
X_ = data_set[:, 0:1]
Y_ = data_set[:, 1:9]

Xx = []
Y = []

for x in X_:
    Xx.append(x[0] * 1. / 255)

Xx = np.array(Xx)
Y = np.array(Y_)

Y /= div

X = Xx[..., np.newaxis]

print(len(X), X.shape)
print(len(Y), Y.shape)
```

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Activation, Dense, Dropout, MaxPooling2D, ZeroPadding2D
from keras.optimizers import Adam
np.random.seed(720)
```

```python
model = Sequential()
model.add(ZeroPadding2D((1,1),input_shape=(72,128,1)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2,2), strides=(2,2)))

model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2,2), strides=(2,2)))

model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(256, (3, 3), activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(256, (3, 3), activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(256, (3, 3), activation='relu'))
model.add(MaxPooling2D((2,2), strides=(2,2)))

model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(512, (3, 3), activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(512, (3, 3), activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(512, (3, 3), activation='relu'))
model.add(MaxPooling2D((2,2), strides=(2,2)))

model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(512, (3, 3), activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(512, (3, 3), activation='relu'))
model.add(ZeroPadding2D((1,1)))
model.add(Conv2D(512, (3, 3), activation='relu'))
model.add(MaxPooling2D((2,2), strides=(2,2)))

model.add(Flatten())
model.add(Dense(4096, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(4096, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(8, activation='linear'))

model.compile(loss='mean_absolute_error', optimizer=Adam(lr=0.001, decay=1e-4), metrics=['mean_absolute_error'])
```

```python
model.fit(X, Y, epochs=500, batch_size=32, shuffle=True, validation_split=0.2)
```

```python
model.save(os.path.join(data_path, 'models', 'cars'))
model.save_weights(os.path.join(data_path, 'weights', 'cars'))
```
