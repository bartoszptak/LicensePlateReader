```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
import glob

arrays = glob.glob('./data/chars/*.npy')

data_path = 'data'
div = 1

data_set = []
for path in arrays:
    local_set = np.load(os.path.join(path))
    for it in local_set:
        data_set.append(it)
data_set = np.array(data_set)
```

```python
data_set.shape
```

```python
for im in data_set:
    img = cv2.resize(im[0], (int(50/div), int(90/div)))
    im[0] = cv2.equalizeHist(img)
```

```python
X_ = data_set[:, 0:1]

Xx = []

for x in X_:
    Xx.append(x[0] * 1. / 255)

Xx = np.array(Xx)

X = Xx[..., np.newaxis]

print(len(X), X.shape)
```

```python
from keras.utils import to_categorical
Y_ = data_set[:, 1:2]
Yy = []

for y in Y_:
    Yy.append(ord(y[0]))

Yy = np.array(Yy)
Y = to_categorical(Yy)

print(len(Y), Y.shape)
print(Y[0], Yy[0])
```

```python
A = Yy.tolist()
A = dict([(e, A.count(e)) for e in A])

Z = []
for x in A:
    Z.append([chr(x),A[x]])

Z.sort(reverse=False)
print(len(Z))
for z in Z:
    print (z[0], z[1])
```

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Activation, Dense, Dropout
np.random.seed(321)
```

```python
model = Sequential()

model.add(Conv2D(128, (3, 3), input_shape=(int(90/div), int(50/div), 1)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(256, (3, 3), padding='same'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(256, (3, 3), padding='same'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())
model.add(Dense(512))
model.add(Activation('relu'))
model.add(Dropout(0.5))

model.add(Dense(91))
model.add(Activation('softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, Y, epochs=100, batch_size=32, shuffle=True, validation_split=0.3)
```

```python
model.save(os.path.join(data_path, 'models', 'chars'))
model.save_weights(os.path.join(data_path, 'weights', 'chars'))
```
